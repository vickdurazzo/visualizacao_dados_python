# **Datasets**

```python
import seaborn as sns
```

```python
# Carregando dataset padrão que vem no seaborn
df_tips = sns.load_dataset("tips")
```

```python
df_tips.head()
```

# **Visualizando Relações Estatísticas nos Dados**


> A análise estatística é o processo de entender como variáveis de dado se relacionam uma com a outra e como essas relações dependem de outras variáveis. A visualização de dados se torna uma peça essencial nesse processo, pois, quando os dados são visualizados adequadamente, as pessoas conseguem descobrir padrões que indicam relações nos dados



# **Função relplot ()**


*   Combina o **scatter plot** e **line plot**
* Muito útil para ver relações entre duas ou mais variáveis com o uso de parâmetros de semântica no dados




# **Relação de Variáveis com Scatter Plots**

**Valor total x gorjeta**

```python
# Gráfico demonstra correlação positiva entre as variáveis
sns.relplot(x="total_bill",y="tip",data=df_tips)
```

**relação entre total da conta x gorjeta e colorindo os pontos com a variável fumante**

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",data=df_tips)
```

**relação entre total da conta x gorjeta e colorindo os pontos com a variável sexo**

```python
sns.relplot(x="total_bill",y="tip",hue="sex",data=df_tips)
```

**relação entre total da conta x gorjeta, colorindo os pontos com a variável sexo e definindo um estilo também baseado na variável sexo**

```python
sns.relplot(x="total_bill",y="tip",hue="sex",style="sex",data=df_tips)
```

**É possível representar quatro variáveis, mas é preciso cautela, pois, a visualização pode ficar confusa**

```python
sns.relplot(x="total_bill",y="tip",hue="sex",style="smoker",data=df_tips)
```

**Em casos de variáveis numéricas para definir a tonalidade a paleta de cores alterna de forma sequencial**

```python
sns.relplot(x="total_bill",y="tip",hue="size",data=df_tips)
```

**Usando parâmetro size para definir o tamanho de cada ponto baseado em alguma coluna da base de dados**

```python
sns.relplot(x="total_bill",y="tip",size="size",data=df_tips);
```

```python
sns.relplot(x="total_bill",y="tip",size="size",sizes=(15,300),data=df_tips);
```

# **Plotando dados categóricos**
* Quando trabalhamos com variáveis categóricas temos que visualizar dados de forma diferentes.
* O Seaborn fornce o método catplot() qe já contém diversos tipos de gráficos embutidos
* Isso facilita, pois, você pode usar diferentes gráficos usando o mesmo método

```python
sns.catplot(x="sex",kind="count",palette="Set2",data=df_tips);
```

 **Gráfico de barras horizontais usando a coluna *day***

```python
sns.catplot(y="day",kind="count",palette="Set1",data=df_tips);
```

**Scatter plot com dados categóricos**

```python
sns.catplot(x="day",y="total_bill",palette="Set2",data=df_tips)
```

**O parâmetro swarn evita sobreposição de pontos**

```python
sns.catplot(x="day",y="total_bill",kind="swarm",data=df_tips);
```

**O parâmetro hue permite adicionarmos uma terceira variável a nossa visualização**

```python
sns.catplot(x="day",y="total_bill",kind="swarm",hue="sex",palette="Dark2",data=df_tips);
```

**O parâmetro *order* permite alternarmos a ordem padrão das categorias que estão sendo exibidas**

```python
sns.catplot(x="day",y="total_bill",kind="swarm",hue="sex",order=['Sat','Sun','Fri','Thur'],palette="Dark2",data=df_tips);
```

**Invertendo a visualização para plots horizontais**

```python
sns.catplot(x="total_bill",y='day',kind="swarm",data=df_tips)
```

# **Gráficos com Regressão**
* Quando temos muitas variáveis quantitativas em nossos dados é interessante visualizar como estas se relacionam.
* Podemos visualizar essas informações com linhas de regressão.
* Com modelos de regressão simples é possível checar se existe alguma **correlacão** entre algumas variáveis.

```python
sns.lmplot(x="total_bill",y="tip",hue="sex",palette="Pastel1",data=df_tips)
```

** Usando o parâmetro col para segregar os gráficos pelo valor da coluna categórica *time* **

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",palette="Set2",data=df_tips)
```

**Parâmetros scatter_kws e line_kws**

```python
sns.lmplot(x="total_bill",
           y="tip",
           hue="smoker",
           col="time",
           palette="Set1",
           data=df_tips,
           scatter_kws={"s":90,"alpha":0.5},
           line_kws={'lw':4})
```

**Segregando gráficos pelo parâmetro *col* e pelo parâmetro *row***

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",col="time", row="sex", palette="Set2", data=df_tips);
```

# **Visualizando distribuição de observações**
* Um dataset muito grande será difícil de visualizar variáveis categóricas.
* É preciso sumarizar a distribuição dos dados para facilitar a visualização.
* Podemos usar um gráfico do tipo Boxplot para visualizar a distribuição de tais variáveis.
* Esse tipo de gráfico é muito útil para visualizar ***Outliers***.

**Boxplot de Dias da semana por total de conta**

```python
sns.catplot(x="day",y="total_bill",kind="box",data=df_tips)
```

**Gráfico de Boxplot dos dias por total de conta e se a pessoa é fumante ou não**

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=df_tips);
```

* Tipo de Boxplot com foco maior na forma dos dados do que nos outliers.
* Esse tipo de gráfico é interessante quando temos grandes datasets.

```python
sns.catplot(x="day", y="total_bill", kind="boxen", data=df_tips);
```

* O Método stripplot permite plotar a distribuição dos dados.
* Podemos combinar os dois gráficos para ter mais informação.

```python
sns.catplot(x="day", y="total_bill", kind="boxen", data=df_tips);
sns.stripplot(x="day", y="total_bill", data=df_tips, color="blue");
```

**O método catplot() permite usarmos facetgrid assim podemos combinar gráficos em uma única figura**

```python
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=df_tips, kind="boxen",height=4);
```

# **Violin Plot**


*   Gráfico que é ideal para visualizar a distribuição de variáveis
*   Cominação do Boxplot com o KDE
* Funciona bem para distribuições com picos
* Permite uma visualização mais rica do que com o Boxplot normal



**Violin Plot**

```python
sns.catplot(x="day",y='total_bill',kind='violin',data=df_tips)
```

**Violin Plot horizontal**

```python
sns.catplot(y="day",x='total_bill',kind='violin',data=df_tips)
```

**Usando o parâmetro Hue para ter uma terceira variável**

```python
sns.catplot(x="day",y='total_bill',hue="time",kind='violin',data=df_tips)
```



*   É possível dividir o violin plot com o parâmetro split
*   Isso é útil quando temos muitos dados para exibir



```python
sns.catplot(x="day",y='total_bill',hue="sex",split=True,kind='violin',data=df_tips)
```



*  Usando o parametro inner para preencher o violin com observações
*  Útil para destacar bem as partes que possuem mais dados



```python
sns.catplot(x="day",y='total_bill',hue="sex",split=True,inner="stick",palette="pastel",kind='violin',data=df_tips)
```

* Podemos combinar também um gráfico de swarm com o Violin Plot
* Dessa froma fica mais explicito ainda a frequencia de pontos nas partes do "violino"

```python
sns.catplot(x="day",y='total_bill',palette="pastel",kind='violin',data=df_tips)
sns.swarmplot(x="day",y="total_bill",color="k",size=3,data=df_tips)
```

# **Joinplot (ScatterPlot e Histograma)**
* Quando queremos ver a distribuição de duas variáveis podemos usar o joinplot
* União de gráficos do tipo Scatter Plot e Histogramas para as duas variáveis
* Mostra o Scatter plot para as duas variáveis e um histograma para cada variável separadamente

```python
sns.jointplot(x="tip",y="total_bill",color='r',data=df_tips)
```

# **Hexbin Plots**
* Adequado para visualização de grande conjuntos de dados
* Tende a funcionar melhor com as cores em branco de fundo

```python
sns.jointplot(x="tip",y="total_bill",color='b',kind="hex",data=df_tips)
```

**Usnado o KDE para plotar mais de uma variável**

```python
sns.jointplot(x="tip",y="total_bill",color='b',kind="kde",data=df_tips)
```

# **Gráficos de Correlação**
* Útil para visualizar se existem correlações positivas ou negativas entre colunas
* Método corr() do pandas possibilita calcular correlação por tipo **spearman** ou **pearson**

```python
correlacoes = df_tips.corr()
```

```python
ax = sns.heatmap(correlacoes)
```

**Exibe os valores de correlacao**

```python
ax = sns.heatmap(correlacoes,annot=True)
```

**Visualizando a correlação de colunas do Dataframe iris**

```python
correlacoes = iris.corr()
```

```python
ax = sns.heatmap(correlacoes,cmap='PuOr',annot=True)
```
